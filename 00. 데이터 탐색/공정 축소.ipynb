> ### 공정명 정규화

1. 용어 통일

SPEED > SPD    
MOTOR > MTR   
DRIVE > DRV

```python
norm_process_df = process_df.copy()
```

```python
def unify_process_names(name):
    name = name.replace('SPEED', 'SPD')
    name = name.replace('MOTOR', 'MTR')
    name = name.replace('DRIVE', 'DRV')
    return name

# 적용 예시: process_df 컬럼명 변경
norm_process_df.columns = [unify_process_names(col) for col in norm_process_df.columns]
print(norm_process_df.columns)
```

2. 점(.) 삽입 - 영어 + 숫자 + (. 삽입) +  영어

Z35TMP > Z35.TMP

```python
import re

def insert_dot_between_zone_and_suffix(s):
    # 1. 영문 + 숫자 + 영문 → 점 삽입 (예: ZONE10PRESSURE → ZONE.PRESSURE)
    s = re.sub(r'([A-Za-z]+)\d+([A-Za-z]{2,})', r'\1.\2', s)
    # 2. 숫자 + 영문 → 점 삽입 (예: 8BAY → .BAY)
    s = re.sub(r'\b\d+([A-Za-z]{2,})\b', r'.\1', s)
    return s

norm_process_df.columns = [insert_dot_between_zone_and_suffix(col) for col in norm_process_df.columns]

print(norm_process_df.columns)

# norm_process_df.to_csv('../data/norm_factory_glass_2016.csv', index=False)
```

> ### 피쳐 추출

1. 공정명 정리

 1) 공정명 정리 : 형식 = 1 + 알파벳(2~5자) + 숫자(2~5자)          
- 센서명 제거   
- 위치 키워드 + 숫자 > 숫자 제거   
- 접미사 제거(.PV, .CPV)
          
 2) 센서명 제거한 공정명 매핑을 위해 저장
             
 3) 센서명 제거 후 _ 나 .으로 구분해서 DF 저장 
   
 4) 공정명 확인

```python
import re

# 센서 ID 및 접미사 제거용 정규표현식
sensor_pattern = re.compile(
    r'(1[A-Z]{2,5}\d{5}[A-Z]*|WOS\d+)',  # 센서 ID
    re.IGNORECASE
)

def clean_column_name(col):
    # 1. 공백, 언더바 → 점으로 정규화
    col = col.replace(' ', '.').replace('_', '.')
    col = re.sub(r'\.+', '.', col).strip('.')
    # 2. 센서 ID 제거
    col = sensor_pattern.sub('', col)
    col = re.sub(r'\.+', '.', col).strip('.')
    # 3. 위치 키워드 + 숫자 → 숫자 제거
    col = re.sub(r'\b(ZONE|Zone|POSITION|M|Z|BAY|SECT|AB|C)(\d+)', r'\1', col, flags=re.IGNORECASE)    
    col = re.sub(r'\.+', '.', col).strip('.')    
    # 4. .PV, .CPV, .숫자 제거 (접미사)
    col = re.sub(r'\.(PV|CPV|\d+)$', '', col, flags=re.IGNORECASE)
    # 5. 숫자 정리
    col = re.sub(r'\b\d+\b', '', col)
    col = re.sub(r'\.+', '.', col).strip('.')    
    return col

# 분리 함수
def split_and_align(col, max_len):
    parts = re.split(r'[._]', col)
    return [None] * (max_len - len(parts)) + parts  # 오른쪽 정렬

# 원본 컬럼 리스트 예시 (process_df.columns 사용 시 적용)
original_columns = norm_process_df.columns  # 또는 샘플 리스트 직접 넣기

# 정제 및 분할
sensor_cleaned = [clean_column_name(col) for col in original_columns] 
split_cleaned = [re.split(r'[._]', c) for c in sensor_cleaned]
max_len = max(len(x) for x in split_cleaned)

# 우측 정렬
aligned = [split_and_align(c, max_len) for c in sensor_cleaned]

cleaned_process_df = pd.DataFrame(sensor_cleaned)
cleaned_process_df.to_csv('../data/original_process_nm_split.csv', index= False, header= False)

split_process_df = pd.DataFrame(aligned)
split_process_df.to_csv('../data/process_nm_split.csv', index=False, header=False)

# 마지막에 PV, CPV, 1, PV.1, CT
# PV : process value
# CPV : Calculated Process Vale(계산된 공정값 / 보정된 센서값)
# PV.1 : 센서가 여러 개의 채널값을 가질 때(이중 센서, 백업 센서)
# CT : Control Target으로 제어 목표값 / 지령값

# 1TI30101.CT : 이 온도 센서가 따라야 하는 설정 목표값일 수 있음

unique_process = set(sensor_cleaned)
print(unique_process)
# len(unique_process) # 722
```

2. 공정명 평균으로 그룹화

```python
# 중복된 열 평균으로 그룹화

seen = set()
ordered_unique_columns = []
for col in norm_process_df.columns:
    if col not in seen:
        seen.add(col)
        ordered_unique_columns.append(col)

# 2. 컬럼별 평균값 시리즈를 딕셔너리에 저장
merged_columns = {}

for col in ordered_unique_columns:
    same_cols = norm_process_df.loc[:, norm_process_df.columns == col]
    merged_columns[col] = same_cols.mean(axis=1)

# 3. 한 번에 병합
merged_df = pd.concat(merged_columns, axis=1)

# 확인
merged_df.info()
```

3. 최종 데이터프레임 생성
- 불량률 열 결합

```python
time_series_idx = df.index
time_series_idx

# 마지막 불량율 열 결합
df_defect = df.iloc[:, -14:]
# df_defect

final_df = pd.concat([merged_df, df_defect], axis= 1)
final_df.index = time_series_idx

final_df.info()
final_df.head()
```
